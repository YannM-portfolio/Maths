This notebook explores the "random operation" model to solve the 'Le Compte est Bon' problem. A statistic study is still left to do, but intuitively, most of the solutions will require less than 50.000 to 100.000 attempts to find.

Let's find a solution to the example in the README file!

```python
nb_list = [3,4,5,2,100,8]
sol = 532
res, operations = random_crack(nb_list, 6, sol)
for op in operations:
    print(op)
```

Now we wonder if there is any solution using less than all 6 numbers

(Spoiler: yes, it is possible to solve the problem using 4 numbers, but not less!)

```python
nb_list = [3,4,5,2,100,8]
sol = 532
res, operations = random_crack(nb_list, 4, sol) # here we modify "6" numbers to "4"
for op in operations:
    print(op)
```

Let's see the concentration of reachable numbers using a given number list and a given number of numbers to use from it!

```python
import matplotlib.pyplot as plt

nb_list = [10,3,7,75,2,1]
nb_to_use = 6
max_iteration = 5*10**4
reach = reachable_target(nb_list, nb_to_use, max_iteration, limit=2000)

plt.hist(reach, bins=max(reach), density=True)
plt.xlim((0, max(reach)+10))
plt.xlabel("Reachable solutions")
plt.title(f"Concentration of reachable solutions using {nb_to_use} numbers")
plt.show()
```
